```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

filename = 'lissajous-300pts.csv'
df = pd.read_csv(filename)
```

```python
x = df.loc[:,'Lx(t)'].values
y = df.loc[:,'Ly(t)'].values
```

```python
plt.style.use('seaborn')
fig, axes = plt.subplots(figsize=(4,5), nrows=1,ncols = 1) 
axes.set_title('Lissajous Curve:', fontsize=22, fontweight='bold', y=1.075)
fig.suptitle(r'(Lx(t),Ly(t))=(sin 5t, sin 4t)', fontsize=14, y=0.935)
axes.plot(x,y, c = '#18006A')
```
